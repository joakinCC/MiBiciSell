# MiBici

MiBici es un sistema de transporte público creado por el gobierno de Jalisco donde puedes tomar y/o dejar una bicicleta.
El sistema cuenta con una red de 236 bahías alrededor de 2000 bicicletas y más de 50,000 usuarios.
Dada la información anterior he pensado que sería un excelente negocio para el 2019 vender aguas frescas a los ciclistas.

Para lograr mi objetivo necesito conocer el comportamiento de los usuarios de MiBici así como la temperatura para poder determinar los puntos en los cuales pueda tener una mayor venta.

```python
import pandas as pd
import folium
```

# DATOS

## Estaciones MiBici

```python
MiBici_Estaciones = pd.read_csv(r'C:/Users/CJO2MX/Documents/MiBiciSell/Data/MiBici/nomenclaturaMiBici.csv', encoding='latin-1')
MiBici_Estaciones.head()
```

Dentro del dataframe vemos que en la columan status solo nos importan aquellas estaciones "IN_SERVICE"

```python
MiBiciEstaciones = MiBici_Estaciones[MiBici_Estaciones.status == 'IN_SERVICE']
```

```python
MiBiciEstaciones = MiBiciEstaciones[['id','name','latitude', 'longitude']]
MiBiciEstaciones = MiBiciEstaciones.reset_index(drop=True)
```

```python
MiBiciEstaciones.head(200)
```

## Mapa

```python
locations = MiBiciEstaciones[['latitude', 'longitude']]
locationlist = locations.values.tolist()
```

```python
map = folium.Map(location=[20.6871392,-103.3881409], zoom_start=12)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=MiBiciEstaciones['name'][point]).add_to(map)
map
```
