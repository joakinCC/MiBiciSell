# Datos de temperatura

A fin de obtener los datos de los contaminantes utilizaremos la API de datos abiertos para lo cual necesitamos instalar el cliente de python. 
https://github.com/mxabierto/api_python_client

```python
from datosgobmx import client

# La fuente será Sinaica
parametros_request = client.makeCall('sinaica-parametros')

#List comprehension
valid_params = [pd.DataFrame.from_dict(v, orient='index').T for v in parametros_request['results']]
valid_params = pd.concat(valid_params, ignore_index=True)
valid_params['date-insert'] = pd.to_datetime(valid_params['date-insert'])
valid_params
```

En parametros que nos proporciona la API viene la temperatura (TMP) reportada en grados centígrados.

```python
d = client.makeCall('sinaica',{'pageSize':5,'state':'Jalisco','parametro':'TMP'})
#List comprehension
valid_param = [pd.DataFrame.from_dict(v, orient='index').T for v in d['results']]
valid_param = pd.concat(valid_param, ignore_index=True)
valid_param['date-insert'] = pd.to_datetime(valid_params['date-insert'])
valid_param
```
